```python
import numpy as np
import matplotlib.pyplot as plt
```

Step 1 - Plot the constraints on a chart to determine the feasible area.

```python
# s + t <= 80
s1 = [80,0]
t1 = [0,80]

# 2s + t <= 100
s2 = [50, 0]
t2 = [0, 100]

# s < 40
s3 = [40, 40]
t3 = [0, 100]
```

```python
plt.plot(s1, t1)
plt.plot(s2, t2)
plt.plot(s3, t3)
```

Step 2 - Determine the optimal solution within the feasible area.

Step 2.1 - Pick a random point within the feasible area and draw a line for the objective function with this point.

We will go with the point <30, 0>.

Our objective function: d = 3s + 2t

This gives us: d = 3 * 30 + 2 * 0 = 90

Which can be rewritten as: t = 45 - 3/2 * s

From which we can say the slope of the objective function is -3/2

```python
plt.plot(s1, t1)
plt.plot(s2, t2)
plt.plot(s3, t3)

# formula of line: y = mx + b
# where m is the slope and b is the y-intercept.
plt.plot([-3/2 * s + 45 for s in np.arange(0, 30)])
```

Step 2.2 - Find the optimal solution by generating lines with increased values of d (objective function). 

The last line within the feasible area will have the maximum value of d. Which will be our answer.

In this example the optimal solution is <20, 60>

```python
plt.plot(s1, t1)
plt.plot(s2, t2)
plt.plot(s3, t3)


plt.plot([-3/2 * s + 90 for s in np.arange(0, 60)])
```
